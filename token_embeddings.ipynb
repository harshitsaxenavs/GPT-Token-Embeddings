```python
# 2. Import modules
from transformers import GPT2Tokenizer, GPT2Model
import torch
import pandas as pd
```

```python
# 3. Load tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2Model.from_pretrained("gpt2")

```

```python
# 4. Input sentence
sentence = input("Enter a sentence: ")

```

```python
# 5. Tokenize input
inputs = tokenizer(sentence, return_tensors="pt")
input_ids = inputs["input_ids"][0]                    # Tensor of token IDs
tokens = tokenizer.convert_ids_to_tokens(input_ids)  # List of tokens

```

```python
# 6. Get embeddings
with torch.no_grad():
    outputs = model(**inputs)
    last_hidden_states = outputs.last_hidden_state  # [1, seq_len, 768]
    embeddings = last_hidden_states[0].numpy()      # [seq_len, 768]


```

```python
# 7. Prepare column names
dimension_cols = [f"Dim_{i}" for i in range(768)]

```

```python
# 8. Create DataFrame with embeddings and tokens
df = pd.DataFrame(embeddings, columns=dimension_cols)
df.insert(0, "Token ID", input_ids.tolist())  # Insert token IDs as first column
df.insert(1, "Token", tokens)                 # Insert tokens after token IDs

```

```python
# 9. Save to CSV
df.to_csv("token_embeddings_full.csv", index=False)
print("\n✅ Saved full embeddings with token and token ID to 'token_embeddings_full.csv'")
```
